```python
from tensorflow.keras.models import load_model

srgan_model = load_model('/kaggle/input/scrnn-66-model/deeper_srcnn_model.h5')  # Update with your actual path
```

```python
import cv2
import numpy as np

def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  
    img = img / 255.0  
    img = np.expand_dims(img, axis=0)  
    return img

def upscale_image(model, image):
    generated_image = model.predict(image)
    generated_image = np.clip(generated_image, 0.0, 1.0)  
    generated_image = (generated_image * 255.0).astype(np.uint8)  
    return generated_image
```

```python
input_image_path = '/kaggle/input/image-pro/Image_Processing/Pixelated/30242.jpg'  
input_image = preprocess_image(input_image_path)
high_res_image = upscale_image(srgan_model, input_image)


cv2.imwrite('output_high_res_image5.jpg', cv2.cvtColor(high_res_image[0], cv2.COLOR_RGB2BGR))
```
